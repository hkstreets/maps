```python
result_obj = {
    "type": "FeatureCollection",
    "features": []
}
```

```python
with open('places.json', 'r') as f:
  data = json.load(f)['USE THIS']
```

```python
data[0]
```

```python
features = []
for place in data:
    obj = {
    "type": "Feature",
    "id" : place['id'],
    "geometry": {
       "type": "Point",
       "coordinates":  [ place['street_long'],place['street_lat'] ]
    },
    "properties": {
    "place": place['place'],
    "country": place['country'],
    "region": place['region'],
    "image": f"https://raw.githubusercontent.com/hkstreets/maps/main/content/projects/asia-world-city/images/{place['id']}.jpg",
    "title": f"{place['street_eng']} {place['street_chi']}" if place['street_chi'] else place['street_eng'],
    "description": place['description'],
    }
  }
    features.append(obj)
```

```python
json_str = json.dumps(features, indent=2)
print(json_str)
```

```python
result_obj['features'] = sorted(features, key=lambda x: x['id']) 
```

```python
result_obj
```

```python
with open('places.geojson', 'w') as f:
    json.dump(result_obj, f, indent=4)
```
